```python
4)

import pandas as pd

def add_word_count_column(df):
    # Split the text in each row by whitespace and count the number of words
    word_counts = df['Text'].str.split().apply(len)
    
    # Add the word_counts as a new column in the DataFrame
    df['Word_Count'] = word_counts
    
    return df

# Example DataFrame
df = pd.DataFrame({'Text': ['This is a sample text.', 'Another example', 'Just a few words']})

# Call the function to add the 'Word_Count' column
df = add_word_count_column(df)

# Print the updated DataFrame
print(df)
```

```python
5)

The DataFrame.size() and DataFrame.shape() methods in Pandas are used to obtain different pieces of information about a DataFrame.

DataFrame.size(): This method returns the total number of elements in the DataFrame. It calculates the size by multiplying the number of rows (DataFrame.shape[0]) by the number of columns (DataFrame.shape[1]). The size() method provides the count of all the cells in the DataFrame, including both the empty and non-empty cells.

DataFrame.shape(): This method returns a tuple representing the dimensions of the DataFrame. It provides information about the number of rows and columns in the DataFrame. The shape is returned as (n_rows, n_columns), where n_rows represents the number of rows in the DataFrame, and n_columns represents the number of columns.

import pandas as pd

data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}

df = pd.DataFrame(data)

print(df.size)     # Output: 9
print(df.shape)    # Output: (3, 3)
```

```python
7)

import pandas as pd

def extract_username(df):
    # Extract the username from the 'Email' column
    df['Username'] = df['Email'].str.split('@').str[0]
    
    return df


# Example DataFrame
df = pd.DataFrame({'Email': ['john@example.com', 'jane@example.com', 'smith@example.com']})

# Call the function to extract the username
df = extract_username(df)

# Print the updated DataFrame
print(df)
```

```python
8)

import pandas as pd

def select_rows(df):
    # Select rows based on conditions
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    
    return selected_rows

# Example DataFrame
df = pd.DataFrame({'A': [3, 8, 6, 2, 9],
                   'B': [5, 2, 9, 3, 1],
                   'C': [1, 7, 4, 5, 2]})

# Call the function to select rows
selected_df = select_rows(df)

# Print the selected DataFrame
print(selected_df)
```

```python
9)

import pandas as pd
import numpy as np

def calculate_statistics(df):
    # Calculate mean, median, and standard deviation
    mean_value = np.mean(df['Values'])
    median_value = np.median(df['Values'])
    std_value = np.std(df['Values'])
    
    # Create a dictionary to store the statistics
    statistics = {
        'Mean': mean_value,
        'Median': median_value,
        'Standard Deviation': std_value
    }
    
    return statistics

# Example DataFrame
df = pd.DataFrame({'Values': [5, 10, 15, 20, 25]})

# Call the function to calculate statistics
statistics = calculate_statistics(df)

# Print the calculated statistics
for stat, value in statistics.items():
    print(f"{stat}: {value}")
```

```python
10)

import pandas as pd

def calculate_moving_average(df):
    # Sort the DataFrame by date in ascending order
    df = df.sort_values('Date')
    
    # Calculate the moving average using a rolling window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df

# Example DataFrame
df = pd.DataFrame({
    'Sales': [10, 15, 8, 12, 20, 18, 25, 21, 16, 22],
    'Date': pd.date_range(start='2023-06-01', periods=10)
})

# Call the function to calculate the moving average
df = calculate_moving_average(df)

# Print the updated DataFrame
print(df)
```

```python
11)

import pandas as pd

def add_weekday_column(df):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Extract weekday name and assign to 'Weekday' column
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df

# Example DataFrame
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})

# Call the function to add the 'Weekday' column
df = add_weekday_column(df)

# Print the modified DataFrame
print(df)
```

```python
12)

import pandas as pd

def select_rows_between_dates(df):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Define the start and end dates
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2023-01-31')
    
    # Select rows between the start and end dates
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows

# Example DataFrame
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-10']})

# Call the function to select rows between dates
selected_df = select_rows_between_dates(df)

# Print the selected DataFrame
print(selected_df)
```
