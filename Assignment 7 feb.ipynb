```python
ANS 1)

import re

def passwordcheck(passw):
    if len(passw) < 10:
        return "invalid password"
    else:
        pass

    upper = 0
    lower = 0
    
    for i in passw:
        if i.isupper():
            upper += 1
        elif i.islower():
            lower += 1
    if upper < 2 or lower < 2:
        return "invalid password"

    if not re.search("[0-9]",passw) or len(re.search("[A-Za-z0-9]",passw)) < 3:
        return "invalid password"

    return "Valid password"


a = input("Enter your password: ")
print(passwordcheck(a))
```
