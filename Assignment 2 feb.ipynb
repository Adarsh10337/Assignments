```python
Ans 1) For loop should be used when we are dealing with loop values in range for example-: 
    for i in range(10,30):
        print("something")
        
    whereas, While should be used when we have to check a specific condition for further implementation of code, for example-:
        while a>0:
            if  a==3:
                print("this")
            else:
                print("this one")
        #Also many conditions can be added
```

```python
ANS 2) For Loop for finding sum of first 10 numbers.
--------------------------------------------------------------

a = 0
for i in range(1,11):
    a =  a + i
print(a)

While loop for finding sum of first 10 numbers.
-----------------------------------------------------------------
num = 10

if num < 0:
   print("Please enter a positive number")
else:
   sum = 0
   while(num > 0):
       sum += num
       num -= 1
   print("The result is", sum)

For loop for multiplication of first 10 numbers
---------------------------------------------------------------

a = 1

for i in range(1,11):
    a = a * i
print(a)
    
While loop for multiplication of first 10 numbers
-----------------------------------------------------------

num = 10

if num < 0:
   print("Please enter a positive number")
else:
   mul = 0
   while(num > 0):
       mul *= num
       num -= 1
   print("The result is", mul)
```

```python
ANS 3)

#Create a python program to compute the electricity bill for a household.
unit = int(input("ENTER THE AMOUNT OF UNITS CONSUMED :"))

if unit > 0 and unit <= 100:
   s = unit * 4.5
   print("You have to pay:",s,"Rs")

   
elif unit > 100 and unit <= 200:
   t = 100*4.5
   b = unit - 100
   c = t + b*6
   print("You have to pay:",c,"Rs")

   
elif unit > 200 and unit <= 300:
   r = 100 * 4.5
   d = 100 * 6
   e = unit - 200
   f = e * 10
   g = r+d+f
   print("You have to pay:",g,"Rs")

else:
   u = 100 * 4.5
   h = 100 * 6
   i = 100 * 10
   j = unit - 300
   k = j * 20
   l = u+h+i+k
   print("You have to pay:",l,"Rs")
```
