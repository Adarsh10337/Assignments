```python
ANS 2)

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
ANS 3)

App routing in Flask refers to the process of defining URL routes for different views or functions in a Flask application. In other words, it involves mapping URLs to specific functions or methods in your Flask application.

We use app routes in Flask for several reasons:

URL structure: App routes allow us to define the URL structure of our Flask application. This helps us create a more organized and structured application that is easy to navigate and understand.

Navigation: App routes allow users to navigate through the different views and pages of our Flask application. By defining app routes, we can create links between different views and allow users to move between them.

Separation of concerns: App routes help to separate the concerns of our Flask application. By mapping specific URLs to specific functions or views, we can create a more modular and maintainable application.

RESTful APIs: App routes are essential for creating RESTful APIs in Flask. By defining specific routes for each API endpoint, we can create a clean and consistent API that is easy to use and understand.
```

```python
ANS 5)

In Flask, the url_for() function is used for URL building. It generates a URL for the specified endpoint and any arguments provided. The function takes the endpoint name as the first argument and any additional arguments as keyword arguments.


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/profile/<username>')
def profile(username):
    return 'Welcome to your profile, {}!'.format(username)

@app.route('/blog/<int:post_id>')
def blog(post_id):
    return 'Welcome to blog post {}!'.format(post_id)

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))  # Output: /
        print(url_for('profile', username='John'))  # Output: /profile/John
        print(url_for('blog', post_id=1))  # Output: /blog/1
        print(url_for('about'))  # Output: /about
```
