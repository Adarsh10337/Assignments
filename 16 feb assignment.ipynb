```python
2)

DDL stands for Data Definition Language, which is a subset of SQL (Structured Query Language) used to define and manage the structure of a database. DDL statements are used to create, modify, and delete database objects such as tables, indexes, and constraints.

Here are some commonly used DDL statements and their purposes:

CREATE: The CREATE statement is used to create new database objects such as tables, indexes, and views. For example, to create a new table called "customers" with columns for customer ID, name, and email, you would use the following SQL statement:
    CREATE TABLE customers (
  customer_id INT PRIMARY KEY,
  name VARCHAR(50),
  email VARCHAR(100)
);

DROP: The DROP statement is used to delete existing database objects. For example, to delete the "customers" table, you would use the following SQL statement:
    DROP TABLE customers;

ALTER: The ALTER statement is used to modify the structure of existing database objects such as tables and columns. For example, to add a new column called "phone" to the "customers" table, you would use the following SQL statement:
    
ALTER TABLE customers ADD COLUMN phone VARCHAR(20);

TRUNCATE: The TRUNCATE statement is used to delete all data from a table, but keep the table structure intact. For example, to delete all data from the "customers" table, you would use the following SQL statement:
    
    TRUNCATE TABLE customers;

DDL statements are used to define and manage the structure of a database, including creating new database objects, modifying existing objects, and deleting objects or their data. These statements are essential for defining the schema of a database and ensuring its integrity over time.
```

```python
3)

DML stands for Data Manipulation Language, which is a subset of SQL (Structured Query Language) used to manipulate the data stored in a database. DML statements are used to insert, update, and delete data in database tables.

Here are some commonly used DML statements and their purposes:

INSERT: The INSERT statement is used to add new data to a database table. For example, to add a new record to the "customers" table with values for customer ID, name, and email, you would use the following SQL statement:
    
INSERT INTO customers (customer_id, name, email) VALUES (1, 'John Smith', 'john.smith@example.com');

UPDATE: The UPDATE statement is used to modify existing data in a database table. For example, to update the email address of a customer with ID 1 in the "customers" table, you would use the following SQL statement:
    
UPDATE customers SET email = 'john.smith@gmail.com' WHERE customer_id = 1;

DELETE: The DELETE statement is used to remove data from a database table. For example, to delete a customer with ID 1 from the "customers" table, you would use the following SQL statement:
    
DELETE FROM customers WHERE customer_id = 1;
```

```python
4)

DQL stands for Data Query Language, which is a subset of SQL (Structured Query Language) used to retrieve data from a database. DQL statements are used to select, filter, and sort data from database tables.

Here is an example of a commonly used DQL statement:

SELECT: The SELECT statement is used to retrieve data from a database table. For example, to select all records from the "customers" table, you would use the following SQL statement:

    SELECT * FROM customers;
This statement retrieves all columns and rows from the "customers" table. You can also use the SELECT statement to retrieve specific columns or rows by specifying the column names or filtering conditions in the query.

For example, to retrieve only the name and email columns for customers with an ID greater than 5, you would use the following SQL statement:
    SELECT name, email FROM customers WHERE customer_id > 5;

This statement retrieves only the "name" and "email" columns from the "customers" table for customers with an ID greater than 5.

You can also use the SELECT statement to sort the results of a query by one or more columns. For example, to retrieve all customers from the "customers" table sorted by name in ascending order, you would use the following SQL statement:
    
    SELECT * FROM customers ORDER BY name ASC;
```

```python
5)

n relational database systems, a primary key and foreign key are used to establish a relationship between two tables. Here is an explanation of each:

Primary Key: A primary key is a column or combination of columns in a database table that uniquely identifies each row in that table. It is a way of ensuring that each row in the table can be identified and accessed efficiently. Primary keys are used as a reference point for other tables in the database, specifically when establishing relationships between tables.
For example, in a "customers" table, the primary key might be a customer ID number. This ID number is unique for each customer in the table, so it can be used to identify and access that customer's information quickly and easily.

Foreign Key: A foreign key is a column or combination of columns in a database table that references the primary key of another table. It establishes a relationship between two tables, allowing information to be accessed across tables in a database. The foreign key is a way of ensuring that data remains consistent across tables, especially when there are multiple tables in the database.
For example, in an "orders" table, there might be a foreign key that references the customer ID number in the "customers" table. This foreign key ensures that each order is associated with a specific customer, and that the data remains consistent across both tables.
```

```python
7)

The order of execution of SQL clauses in an SQL query is as follows:

FROM: The FROM clause specifies the table or tables that the query is operating on. This clause is executed first, and it determines the set of rows that the query will operate on.

JOIN: If the query involves a join operation, the JOIN clause is executed next. This clause combines rows from multiple tables based on a specified condition.

WHERE: The WHERE clause is executed next. This clause filters the rows from the result set that do not meet a specified condition.

GROUP BY: The GROUP BY clause is executed next. This clause groups the rows in the result set based on a specified column or columns.

HAVING: The HAVING clause is executed next. This clause filters the groups created by the GROUP BY clause based on a specified condition.

SELECT: The SELECT clause is executed next. This clause specifies the columns that should be included in the result set.

DISTINCT: If the query includes the DISTINCT keyword, this clause is executed next. This clause removes duplicates from the result set.

ORDER BY: The ORDER BY clause is executed next. This clause sorts the rows in the result set based on a specified column or columns.

LIMIT/OFFSET: If the query includes the LIMIT and/or OFFSET keywords, these clauses are executed last. These clauses limit the number of rows returned by the query and/or specify a starting point for the result set.
```
