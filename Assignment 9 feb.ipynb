```python
ANS 3)

Multiple inheritance is a feature of object-oriented programming where a class can inherit from more than one base class. This allows a class to inherit and reuse code and data from multiple sources, which can help to make the code more modular and easier to maintain.

Here's an example implementation of a Python code that demonstrates multiple inheritance:

class Animal:
    def speak(self):
        print("Animal speaks")

class Mammal(Animal):
    def run(self):
        print("Mammal runs")

class Bird(Animal):
    def fly(self):
        print("Bird flies")

class Bat(Mammal, Bird):
    def echo_location(self):
        print("Bat uses echo location")

# create an instance of the Bat class
my_bat = Bat()

# call methods on the Bat instance
my_bat.speak() # outputs "Animal speaks"
my_bat.run() # outputs "Mammal runs"
my_bat.fly() # outputs "Bird flies"
my_bat.echo_location() # outputs "Bat uses echo location"


In this example, we define four classes: Animal, Mammal, Bird, and Bat. The Animal class defines a speak method, which prints a message to the console. The Mammal and Bird classes both inherit from Animal, and define run and fly methods respectively. The Bat class inherits from both Mammal and Bird, and defines an additional echo_location method.

We create an instance of the Bat class, and then call methods on the instance to demonstrate that it can access methods from both the Mammal and Bird classes, as well as its own echo_location method. This shows how multiple inheritance can be used to combine functionality from multiple sources to create a more complex and versatile class.
```

```python
ANS 4)

In Python, getter and setter methods are used to access and modify the values of instance variables in a class, respectively. Getter methods are used to retrieve the value of an instance variable, while setter methods are used to change the value of an instance variable.

Here's an example implementation of a Python class that defines a getter and setter method:

class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age
    
    def get_name(self):
        return self._name
    
    def set_name(self, name):
        self._name = name
        
    def get_age(self):
        return self._age
    
    def set_age(self, age):
        self._age = age

In this example, we define a Person class that has two instance variables: _name and _age. We define getter and setter methods for both instance variables. The getter methods simply return the value of the corresponding instance variable, while the setter methods take a new value as an argument and set the corresponding instance variable to the new value.

Note that the names of the getter and setter methods follow a specific convention: the getter method has the prefix "get_" followed by the name of the instance variable, while the setter method has the prefix "set_" followed by the name of the instance variable.

Here's an example usage of the Person class:

# create a new Person object
p = Person("Alice", 30)

# get the current name and age
print(p.get_name()) # outputs "Alice"
print(p.get_age()) # outputs 30

# set a new name and age
p.set_name("Bob")
p.set_age(25)

# get the updated name and age
print(p.get_name()) # outputs "Bob"
print(p.get_age()) # outputs 25


In this example, we create a new Person object with the name "Alice" and age 30. We then use the getter and setter methods to get and set the name and age of the Person object.

```

```python
ANS 5)

Method overriding is a feature of object-oriented programming that allows a subclass to provide its own implementation of a method that is already defined in its superclass. When a method is overridden, the implementation in the subclass replaces the implementation in the superclass.

Here's an example implementation of a Python class that demonstrates method overriding:

class Animal:
    def make_sound(self):
        print("The animal makes a sound.")

class Cat(Animal):
    def make_sound(self):
        print("The cat meows.")

class Dog(Animal):
    def make_sound(self):
        print("The dog barks.")

        
In this example, we define an Animal class with a make_sound method that simply prints a generic message. We then define two subclasses, Cat and Dog, that inherit from Animal. Each subclass provides its own implementation of the make_sound method that is specific to the type of animal.

Here's an example usage of the Animal, Cat, and Dog classes:

# create some Animal objects
a1 = Animal()
c1 = Cat()
d1 = Dog()

# call the make_sound method on each object
a1.make_sound() # outputs "The animal makes a sound."
c1.make_sound() # outputs "The cat meows."
d1.make_sound() # outputs "The dog barks."


In this example, we create an Animal object as well as a Cat and a Dog object. We call the make_sound method on each object, and we see that the implementation that is called depends on the type of object. The make_sound method is overridden in each subclass to provide a specific implementation that is appropriate for that type of animal.
```
