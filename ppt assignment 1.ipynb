```python
def twoSum(nums, target):
    complement_dict = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in complement_dict:
            return [complement_dict[complement], i]
        complement_dict[num] = i
    return []

# Example usage
nums = [2, 7, 11, 15]
target = 9
result = twoSum(nums, target)
print(result)
```

```python
def removeElement(nums, val):
    left = 0
    right = len(nums) - 1

    while left <= right:
        if nums[left] == val:
            nums[left], nums[right] = nums[right], nums[left]
            right -= 1
        else:
            left += 1

    return left

# Example usage
nums = [3, 2, 2, 3]
val = 3
k = removeElement(nums, val)
print("k =", k)
print("nums =", nums[:k])
```

```python
def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left

# Example usage
nums = [1, 3, 5, 6]
target = 5
index = searchInsert(nums, target)
print(index)
```

```python
def plusOne(digits):
    n = len(digits)
    
    # Start from the rightmost digit
    for i in range(n-1, -1, -1):
        # Increment the current digit
        digits[i] += 1
        
        # If the digit is less than 10, no carry needed
        if digits[i] < 10:
            return digits
        
        # Carry over to the next digit
        digits[i] = 0
    
    # If we reach here, a carry is needed at the leftmost digit
    return [1] + digits

# Example usage
digits = [1, 2, 3]
result = plusOne(digits)
print(result)
```

```python
def merge(nums1, m, nums2, n):
    # Initialize two pointers for nums1 and nums2
    p1 = m - 1
    p2 = n - 1
    
    # Set the pointer for the merged array
    p = m + n - 1
    
    # Merge nums1 and nums2 from right to left
    while p1 >= 0 and p2 >= 0:
        if nums1[p1] >= nums2[p2]:
            nums1[p] = nums1[p1]
            p1 -= 1
        else:
            nums1[p] = nums2[p2]
            p2 -= 1
        p -= 1
    
    # Copy any remaining elements from nums2 to nums1
    nums1[:p2 + 1] = nums2[:p2 + 1]

# Example usage
nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3

merge(nums1, m, nums2, n)
print(nums1)
```

```python
def containsDuplicate(nums):
    num_set = set()
    
    for num in nums:
        if num in num_set:
            return True
        num_set.add(num)
    
    return False

# Example usage
nums = [1, 2, 3, 1]
result = containsDuplicate(nums)
print(result)
```

```python
def moveZeroes(nums):
    zero_ptr = 0  # Pointer to track the position to place the next non-zero element
    
    # Move non-zero elements to the left
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[i], nums[zero_ptr] = nums[zero_ptr], nums[i]
            zero_ptr += 1

# Example usage
nums = [0, 1, 0, 3, 12]
moveZeroes(nums)
print(nums)
```

```python
def findErrorNums(nums):
    n = len(nums)
    num_set = set()
    duplicate = -1
    
    # Find the duplicate number
    for num in nums:
        if num in num_set:
            duplicate = num
        else:
            num_set.add(num)
    
    # Find the missing number
    missing = -1
    for i in range(1, n+1):
        if i not in num_set:
            missing = i
            break
    
    return [duplicate, missing]

# Example usage
nums = [1, 2, 2, 4]
result = findErrorNums(nums)
print(result)
```
